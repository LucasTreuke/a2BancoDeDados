```python
import geopandas
import pandas as pd

filename = "countries.geojson"
file = open(filename)
df = geopandas.read_file(file)


df.plot()
```

```python
df.rename(columns={'sovereignt':'Country'}, inplace=True)
df.head()
```

```python
filename2 = "HDI.csv"
file2 = open(filename2)
df2 = pd.read_csv(file2)
df2.head(8)
df3 = pd.merge(df,df2,on="Country")
df3.plot()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)

df = df[df["ESTUF"]!="SC"]
df.plot(column="ALCANCE", ax=ax, legend=True)
```
