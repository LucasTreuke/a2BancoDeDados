```python
import os
import sys
import time
import datetime
import numpy as np
import pandas as pd

from tabulate import tabulate

import sqlite3
```

```python
df = pd.read_csv("cursos-prouni.csv")
df.columns
```

```python
conn = sqlite3.connect(os.path.join('cursos_prouni.sqlite'))
cur = conn.cursor()
```

```python
cur.execute("CREATE TABLE IF NOT EXISTS cursos_prouni ('grau', 'turno', 'mensalidade', 'bolsa_integral_cotas','bolsa_integral_ampla', 'bolsa_parcial_cotas', 'bolsa_parcial_ampla','curso_id', 'curso_busca', 'cidade_busca', 'uf_busca', 'cidade_filtro','universidade_nome', 'campus_nome', 'campus_id', 'nome','nota_integral_ampla', 'nota_integral_cotas', 'nota_parcial_ampla','nota_parcial_cotas')")
```

```python
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(tabulate(cur.fetchall()))
```

```python
cur.execute("SELECT grau FROM cursos_prouni;")
print(tabulate(cur.fetchall()))
```

```python
df.to_sql('cursos_prouni', conn, if_exists='replace', index = False)
```

```python
conn.commit()
```
